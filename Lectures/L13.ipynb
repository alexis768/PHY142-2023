### Wheatstone Bridge

![im.png](images/wheatstone.png)

The circuit we have just seen is in fact the <a href="https://en.wikipedia.org/wiki/Wheatstone_bridge" class="interwiki iw_wp" title="https://en.wikipedia.org/wiki/Wheatstone_bridge">Wheatstone Bridge</a>. With a Wheatstone bridge the value of a unknown resistor $R_{x}$ can be measured to high precision by the use of an adjustable resistor, $R_{3}$. The idea is that when there is zero potential measured across the voltmeter in the middle of the bridge, the values of $R_{3}$ and $R_{x}$ are the same. The equations that explain the operation of the bridge can be derived from Kirchoff's rules.

$I_{1}+I_{G}=I_{2}$
    
$I_{3}=I_{G}+I_{x}$
    
$I_{3}R_{3}+I_{G}R_{G}-I_{1}R_{1}=0$
    
$I_{G}R_{G}+I_{2}R_{2}-I_{x}R_{x}=0$

If we balance the bridge then $I_{G}=0$ so the  equations become

$I_{1}=I_{2}$

$I_{3}=I_{x}$

$I_{3}R_{3}=I_{1}R_{1}$

$I_{x}R_{x}=I_{2}R_{2}$

$\frac{I_{x}R_{x}}{I_{3}R_{3}}=\frac{I_{2}R_{2}}{I_{1}R_{1}}$

$R_{x}=\frac{I_{2}R_{2}I_{3}R_{3}}{I_{1}R_{1}I_{x}}\to=\frac{R_{2}R_{3}}{R_{1}}$


### Meters

One reason for using a Wheatstone bridge is that it removes any effects that can arise to non-idealities in meter characteristics. When we want to measure something we want to have as little effect on the measurement as possible.

When using a voltmeter we want to measure the potential between two points, but without having any current flow through the voltmeter. By contrast, we place an ammeter in series with the device we want to measure, and want the potential drop across it to be as small as possible. Therefore the resistance of a voltmeter should be as high as possible, and the resistance of an ammeter should be as low as possible. 

![im.png](images/ohmslaw_fig1.gif)


### Making Meters

Both meters can be assembled by combining a <a href="https://en.wikipedia.org/wiki/Galvanometer" class="interwiki iw_wp" title="https://en.wikipedia.org/wiki/Galvanometer">Galvanometer</a> with resistors. A Galvanometer is a meter which measures a small current based on electromagnetic induction (we'll come back to this later!). For an ammeter a shunt resistor is placed in parallel to allow most of the current to flow through the shunt. For a voltmeter a resistor is placed in series to reduce the total amount of current that flows through the meter. As the resistor of the galvanometer itself is known the actual current or voltage can be worked out  (the meter scale usually takes care of this automatically!).

### Linear algebra for solving circuits

The application of Kirchoff&#039;s laws eventuates in a <a href="https://en.wikipedia.org/wiki/System_of_linear_equations" class="interwiki iw_wp" title="https://en.wikipedia.org/wiki/System_of_linear_equations">system of linear equations</a> we need to solve. As system&#039;s become increasingly complicated the work in solving these equations as we have done earlier becomes tedious and painful. We should note however that we can write a set of equations in terms of matrices and vectors and use linear algebra techniques to accelerate our problem solving).
</p>

<p>
<a href="/~mdawber/_detail/phy142:lectures:kirchoff2b.png?id=phy142%3Alectures%3A13" class="media" title="phy142:lectures:kirchoff2b.png"><img src="/~mdawber/_media/phy142:lectures:kirchoff2b.png" class="media" alt="" /></a>
</p>

<p>
For example, the circuit with two sources of emf we looked at earlier was characterized by the 3 equations
</p>

<p>
$\mathcal{E}_{1}-I_{3}R_{3}-I_{1}R_{1}=0$<br/>

$\mathcal{E}_{2}-I_{3}R_{3}-I_{2}R_{2}=0$<br/>

$I_{1}+I_{2}=I_{3}$
</p>

<p>
A small rearrangement gives us
</p>

<p>
$R_{1}I_{1}+R_{3}I_{3}=\mathcal{E}_{1}$<br/>

$R_{2}I_{2}+R_{3}I_{3}=\mathcal{E}_{2}$<br/>

$I_{1}+I_{2}-I_{3}=0$
</p>

<p>
which can be written as
</p>

<p>
$$\left[ \begin{array}{ccc}
R_{1} ; 0 ; R_{3} \\
0 ; R_{2} ; R_{3} \\
1 ; 1 ; -1 \end{array} \right]  \left[ \begin{array}{c}
I_{1}\\
I_{2} \\
I_{3} \end{array} \right]= \left[ \begin{array}{c}
\mathcal{E}_{1}\\
\mathcal{E}_{2} \\
0 \end{array} \right]$$
</p>

<p>
This is now a <a href="http://mathworld.wolfram.com/MatrixEquation.html" class="urlextern" title="http://mathworld.wolfram.com/MatrixEquation.html">Matrix equation</a> to which linear algebra techniques can be applied. Perhaps the most attractive feature of Matrix equations is that very efficient computational algorithms exist for solving them.

### Gaussian Elimination</h2>
<div class="level2">

<p>
In Gaussian elimination the idea is to reduce the matrix to <a href="https://en.wikipedia.org/wiki/Row_echelon_form" class="interwiki iw_wp" title="https://en.wikipedia.org/wiki/Row_echelon_form">reduced row echelon form</a> via elementary row operations. You can think of these as the things you are allowed to do to a matrix
</p>
<ol>
<li class="level1"><div class="li"> Swap the positions of two rows.</div>
</li>
<li class="level1"><div class="li"> Multiply a row by a nonzero scalar.</div>
</li>
<li class="level1"><div class="li"> Add to one row a scalar multiple of another. </div>
</li>
</ol>

<p>
These operations are performed on the <a href="https://en.wikipedia.org/wiki/Augmented_matrix" class="interwiki iw_wp" title="https://en.wikipedia.org/wiki/Augmented_matrix">augmented matrix</a> which combines both the left and right hand sides of our system of equations.
</p>

<p>
$$\left[ \begin{array}{ccc}
R_{1} ; 0 ; R_{3} \\
0 ; R_{2} ; R_{3} \\
1 ; 1 ; -1 \end{array} \right]  \left[ \begin{array}{c}
I_{1}\\
I_{2} \\
I_{3} \end{array} \right]= \left[ \begin{array}{c}
\mathcal{E}_{1}\\
\mathcal{E}_{2} \\
0 \end{array} \right] \to \left[ \begin{array}{ccc}
R_{1} ; 0 ; R_{3} \\
0 ; R_{2} ; R_{3} \\
1 ; 1 ; -1 \end{array} \right| \left. \begin{array}{c}
\mathcal{E}_{1}\\
\mathcal{E}_{2} \\
0 \end{array} \right]$$
</p>

<p>
The approach to follow during Gaussian elimination is described <a href="http://mathworld.wolfram.com/GaussianElimination.html" class="urlextern" title="http://mathworld.wolfram.com/GaussianElimination.html">here</a>. However, when solving numerical problems your calculator can do this very effectively for you. For instructions on doing this on the TI-83 <a href="http://www2.cfcc.edu/faculty/cmoore/TI83MatrixOper.htm" class="urlextern" title="http://www2.cfcc.edu/faculty/cmoore/TI83MatrixOper.htm">see here</a>. Of course, when you have access to a computer, python is the easiest way to solve it!

### Example
![im.png](images/numericalkirchoff.png)

$20-I_{3}470-I_{2}3=0$<br/>

$12-I_{3}470-I_{1}2=0$<br/>

$I_{1}+I_{2}=I_{3}$
</p>

<p>
$3I_{2}+470I_{3}=20$<br/>

$2I_{1}+470I_{3}=12$<br/>

$I_{1}+I_{2}-I_{3}=0$
</p>

<p>
$$ \left[ \begin{array}{ccc}
0 ; 3 ; 470 \\
2 ; 0 ; 470 \\
1 ; 1 ; -1 \end{array} \right| \left. \begin{array}{c}
20\\
12 \\
0 \end{array} \right]$$
</p>

```python
import numpy as np

A = [[0,3,470],
    [2,0,470],
    [1,1,-1]]
b = [20,12,0]
# Now simply solve for x
I = np.dot(np.linalg.inv(A), b) 
#np.linalg.inv(A)  is simply the inverse of A, np.dot is the dot product
print(I)
```

### RC circuits

When we consider a circuit with a resistor and capacitor in series with a battery and a switch and consider the case where the capacitor is initially uncharged we can reason that when the switch is closed a current needs to flow from the battery to the capacitor. The current can be analyzed based on the DC circuit concepts we have just been discussing.

![im.png](images/rccharging.png)

From Kirchoff's loop rule

$\mathcal{E}-I(t)R-\frac{Q(t)}{C}=0$

Writing this in terms of charge

$\mathcal{E}-\frac{dQ}{dt}R-\frac{1}{C}Q=0$

and then

$\mathcal{E}C-Q=\frac{dQ}{dt}RC$

$\frac{dQ}{\mathcal{E}C-Q}=\frac{dt}{RC}$

At time $t=0$, $Q=0$ and at time $t=t$, $Q=Q$

$\int_{0}^{Q}\frac{dQ}{\mathcal{E}C-Q}=\int_{0}^{t}\frac{dt}{RC}$

$[-\ln(\mathcal{E}C-Q)]_{0}^{Q}=[\frac{t}{RC}]_{0}^{t}$

$\ln(\frac{\mathcal{E}C}{\mathcal{E}C-Q})=\frac{t}{RC}$

$\ln(\frac{\mathcal{E}C-Q}{\mathcal{E}C})=-\frac{t}{RC}$

$\frac{\mathcal{E}C-Q}{\mathcal{E}C}=e^{-\frac{t}{RC}}$


$Q=\mathcal{E}C(1-e^{-\frac{t}{RC}})$


The voltage on the capacitor is given by $V_{C}=\frac{Q}{C}$


$V_{C}=\mathcal{E}(1-e^{-\frac{t}{RC}})$


$I=\frac{dQ}{dt}=\frac{\mathcal{E}}{R}e^{-\frac{t}{RC}}$
